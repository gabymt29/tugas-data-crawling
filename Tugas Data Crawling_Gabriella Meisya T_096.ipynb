```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
import time
```

```python
#INSTALL BROWSER SUCH CHROME OR EDGE
chromedriver_autoinstaller.install()

#USE THE INSTALLED BROWSER FOR SELENIUM
browser = webdriver.Chrome()
```

```python
data = []

for index in range (1,36):
    browser.get(f'https://www.carsome.id/beli-mobil-bekas?pageNo={index}')
    time.sleep(10)
    soup = BeautifulSoup(browser.page_source, 'lxml')
    rows = soup.find_all('div', class_='list-card__item')
    
    for row in rows:
        merk_element = row.find(class_='mod-card__title')
        merk = merk_element.text.replace("\n", "").strip() if merk_element is not None and merk_element.text.strip() != "" else None
        price_total_element = row.find(class_='mod-card__price__total')
        price_total = price_total_element.text.replace("\n", "").strip() if price_total_element is not None and price_total_element.text.strip() != "" else None
        price_cash_element = row.find(class_='mod-card__price-cash')
        price_cash = price_cash_element.text.replace("\n", "").strip() if price_cash_element is not None and price_cash_element.text.strip() != "" else None
        price_month_element = row.find(class_='mod-tooltipMonthPay')
        price_month = price_month_element.text.replace("\n", "").strip() if price_month_element is not None and price_month_element.text.strip() != "" else None
        keterangan_element = row.find(class_='mod-card__car-other')
        keterangan = keterangan_element.text.replace("\n", "").strip() if keterangan_element is not None and keterangan_element.text.strip() != "" else None
    
        dt = {
            'merk': merk,
            'price_total': price_total,
            'price_cash': price_cash,
            'price_month': price_month,
            'keterangan' : keterangan
            }
        
        data.append(dt)
    
    
```

```python
df = pd.DataFrame(data)
```
